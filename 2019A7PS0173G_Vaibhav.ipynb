```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(1000,1000))
plt.imshow(blank_img,cmap='gray')
```

```python
blank_img
```

```python
ix=500
iy=500
gradient_count=0
area_count=500

while gradient_count <=255 :
    blank_img[ix-area_count:ix+area_count,iy-area_count:iy+area_count]=gradient_count

    gradient_count=gradient_count+1
    area_count=area_count-5

plt.imshow(blank_img,cmap='gray')
```
