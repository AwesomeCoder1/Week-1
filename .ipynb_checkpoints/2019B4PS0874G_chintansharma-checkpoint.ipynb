```python
import numpy as np
from skimage import io
from PIL import Image
photo= io.imread('download1.jpg') # for inverted image, inverted color image continued at the end
io.imshow(photo[::-1])
```

```python
arr=np.array(photo)# converting photo into array
arr
```

```python
img=Image.fromarray(arr)# for converting array into photo and saving it as another file
img.save('my.png')
img.show()
```

```python
gray= np.dot(arr[...,:3],[0.2989, 0.5870, 0.1140])# for grayscale image
io.imshow(gray,cmap="gray")
```
