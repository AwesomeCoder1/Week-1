```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tri = np.random.default_rng().triangular(-3, 0, 3, (1000,1000))

tri.sort()
```

```python
def pdf(y):
    arr=[]
    for x in y:
        if (x>=-3 and x<=0):
            arr.append((x+3)/9)
        elif (x>0 and x<=3):
            arr.append((3-x)/9)
        else :
            arr.append(0)
    return arr
```

```python
tri_pdf = np.apply_along_axis(pdf, 0, tri)
```

```python
tri_pdf
```

```python
plt.imshow(tri_pdf*tri_pdf.T, cmap=plt.cm.binary_r)
```
